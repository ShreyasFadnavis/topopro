Jupyter notebook to generate surface plots for the manuscript
===

#### NOTE: In order to run this notebook start with the first and very last cell (so that ivim may be redefined without cluter in the visual ouputs).

```python
# Imports
import time
import matplotlib.pyplot as plt
import numpy as np
import hyperct
from dipy.reconst.ivim import IvimModel  # REDEFINED LATED 
from dipy.data.fetcher import read_ivim
```

```python
# Load data
img, gtab = read_ivim()
data = img.get_data()
print('data.shape (%d, %d, %d, %d)' % data.shape)
```

```python
# Define common functions
def plot_map(raw_data, variable, limits, filename):
    fig, ax = plt.subplots(1)
    lower, upper = limits
    ax.set_title('Map for {}'.format(variable))
    im = ax.imshow(raw_data.T, origin='lower', clim=(lower, upper),
                   cmap="viridis", interpolation='nearest')
    fig.colorbar(im)
    fig.savefig(filename)
```

```python
# Locate the voxels used in the plots
z = 33
b = 0

variable = '$S_0$'
fig, ax = plt.subplots(1)
ax.set_title('Map for {}'.format(variable))
im = ax.imshow(data[:, :, z, b].T, origin='lower', cmap='viridis',
           interpolation='nearest')
fig.colorbar(im)
plt.axhline(y=90, color='k')
plt.axvline(x=90, color='k')
plt.axhline(y=140, color='r')
plt.axvline(x=126, color='r')

plt.axhline(y=135, color='g')
plt.axvline(x=127, color='g')
```

```python
fig, ax = plt.subplots(1)
x1, x2 = 90, 91
y1, y2 = 90, 91
data_slice = data[x1:x2, y1:y2, z, :]
im = ax.imshow(data[x1:x2, y1:y2, z, b].T, origin='lower',
           cmap='viridis', interpolation='nearest')
fig.colorbar(im)
```

```python
ts = time.time()
ivimmodel_vp = IvimModel(gtab, fit_method='VarPro')
ts = time.time()
ivimfit_vp = ivimmodel_vp.fit(data_slice)
```
