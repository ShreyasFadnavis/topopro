```python
import matplotlib.pyplot as plt
import numpy as np
from dipy.reconst.ivim import IvimModel
from dipy.core.gradients import gradient_table
from dipy.data import get_fnames
from dipy.io.gradients import read_bvals_bvecs
from dipy.io.image import load_nifti_data
```

```python
fraw, fbval, fbvec = get_fnames('ivim')
```

```python
data = load_nifti_data(fraw)
bvals, bvecs = read_bvals_bvecs(fbval, fbvec)
gtab = gradient_table(bvals, bvecs, b0_threshold=0)
print('data.shape (%d, %d, %d, %d)' % data.shape)
```

```python
mask, _ = load_nifti('ivim_mask (1).nii.gz')
```

```python
t1 = time.time()
ivimmodel_vp = IvimModel(gtab, bounds=([0., 0., 0.], [0.9, 0.9, 0.9]), fit_method='VarPro')
ivimfit_vp = ivimmodel_vp.fit(data[..., 33, :], mask[..., 33])
t2 = time.time()
print("Time Taken: ", t2 -t1) 
```

```python
# helper function to see plots with matplotlib:
def plot_map(raw_data, variable, limits, filename):
    fig, ax = plt.subplots(1)
    lower, upper = limits
    ax.set_title('Map for {}'.format(variable))
    im = ax.imshow(raw_data.T, origin='lower', clim=(lower, upper),
                   cmap="inferno", interpolation='nearest')
    fig.colorbar(im)
    fig.savefig(filename)
```

```python
plot_map(ivimfit_vp.perfusion_fraction, "f", (0, .5), "perfusion_fraction.png")
plot_map(ivimfit_vp.D_star, "D*", (0., 0.015), "perfusion_coeff.png")
plot_map(ivimfit_vp.D, "D", (0, 0.0015), "diffusion_coeff.png")
```
